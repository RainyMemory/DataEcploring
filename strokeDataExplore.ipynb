````python
feature_list = [['age'], ['avg_glucose_level'], ['bmi']]
data_cols = []
for cols in feature_list:
    for col in cols:
        data_cols.append(data_source[col].dropna())
    plt.figure(figsize=(15,len(cols)))
    plt.boxplot(x=data_cols, vert=False, labels=cols)
    plt.show()
    data_cols = []
````

````python
# Categlorical的数据有：
# 二分类，Binary：'gender'(other类在此样本集中太少，缺少数学价值), 'ever_married', 'Residence_type'
# 顺序型，Ordinal / 多类，Multi-class：'smoking_status'
# 多类，Multi-class：'work_type'
data_source.gender[data_source.gender == 'Male'] = 0
data_source.gender[data_source.gender == 'Female'] = 1
data_source.gender[data_source.gender == 'Other'] = np.nan
data_source.dropna(subset=['gender'], inplace=True)
data_source.ever_married[data_source.ever_married == 'No'] = 0
data_source.ever_married[data_source.ever_married == 'Yes'] = 1
data_source.Residence_type[data_source.Residence_type == 'Rural'] = 0
data_source.Residence_type[data_source.Residence_type == 'Urban'] = 1
````

````python
smoke_df = pd.get_dummies(data_source.smoking_status)
work_df = pd.get_dummies(data_source.work_type)
data_source = pd.concat([data_source, smoke_df, work_df], axis=1)
data_source.columns
````

````python
data_source.drop(columns=['work_type', 'smoking_status'], inplace=True)
print(data_source.size/data_source.columns.size - data_source.count())
````

````python
# age，avg_glucose_level和bmi需要normalise
def do_norm(key_list, norm_df, max_val, min_val):
    norm_temp = []
    for key_temp in key_list:
        for temp in norm_df[key_temp]:
            norm_val = (temp - min_val) / (max_val - min_val)
            if norm_val < 0:
                norm_val = 0
            elif norm_val > 1:
                norm_val = 1
            norm_temp.append(norm_val)
        norm_df['Norm'+key_temp] = pd.Series(norm_temp)
        norm_temp = []
    norm_df.drop(columns=key_list, inplace=True)
````

````python
# age部分
max_age = data_source.age.max()
min_age = 0
do_norm(['age'], data_source, max_age, min_age)

# avg_glucose_level部分
max_glu = data_source.avg_glucose_level.max()
min_glu = data_source.avg_glucose_level.min()
do_norm(['avg_glucose_level'], data_source, max_glu, min_glu)

# bmi部分，关注一下outliers（bar plot右侧离散度过大）
max_bmi = data_source.bmi.quantile(.99)
min_bmi = data_source.bmi.min()
do_norm(['bmi'], data_source, max_bmi, min_bmi)
````
